#### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

#### Line Plot (sns.lineplot)
Use: Line plots are used to display data points and trends over a continuous variable such as time. They are useful for visualizing trends and patterns over time or other ordered sequences.
#### Box Plot (sns.boxplot)
Use: Box plots summarize the distribution of a dataset by showing the median, quartiles, and outliers. They are useful for comparing distributions across different groups or categories.
#### Histogram (sns.histplot)
Use: Histograms display the distribution of a single variable by dividing the data into bins and counting the number of observations in each bin. They are useful for understanding the frequency distribution of a variable.
#### Pair Plot (sns.pairplot)
Use: Pair plots are used to visualize pairwise relationships in a dataset. They create a matrix of scatter plots for each pair of variables in the data and are useful for exploring correlations and interactions between variables.
#### Heatmap (sns.heatmap)
Use: Heatmaps display data in a matrix format using colors to represent values. They are useful for visualizing the magnitude of values across two dimensions and identifying patterns or correlations in the data.

#### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
#### "timepoint" and y = "signal" for different events and regions.
#### Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False)
plt.title('FMRI Signal Over Time')
plt.show()
```

#### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
#### 'pclass', y = 'age' and y = 'fare'.
#### Note: pclass, age, and fare are columns in the titanic dataset.

```python
# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create box plot for age
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Age Distribution by Passenger Class')

# Create box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Fare Distribution by Passenger Class')

plt.tight_layout()
plt.show()
```

#### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
#### parameter for the 'cut' column of the diamonds dataset.

```python
# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Create histogram for price with hue for cut
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple="stack")
plt.title('Price Distribution by Cut')
plt.show()
```

#### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
#### of the iris dataset.

```python
# Load the iris dataset
iris = sns.load_dataset("iris")

# Create pair plot with hue for species
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

#### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the data for the heatmap
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Flights Dataset')
plt.show()
```
