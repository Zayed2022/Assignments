Q1. Explain GET and POST methods.
GET Method:

Definition: The GET method is used to request data from a specified resource.
Usage: It sends data in the URL, and the data is visible to everyone in the URL.
Characteristics: GET requests can be cached, remain in the browser history, and be bookmarked. They should only be used for requests that do not change server state.
Example: Fetching user information from a server.
POST Method:

Definition: The POST method is used to send data to a server to create/update a resource.
Usage: It sends data in the HTTP body and is more secure than GET.
Characteristics: POST requests are not cached, do not remain in the browser history, and cannot be bookmarked.
Example: Submitting user information to a server.


Q2. Why is request used in Flask?
In Flask, the request object is used to access incoming request data, which is data sent to the server via a client request. It allows you to access form data, file uploads, cookies, and request headers. The request object in Flask allows you to handle the data submitted with a POST or GET request.

Q3. Why is redirect() used in Flask?
In Flask, the redirect() function is used to redirect the client to a different endpoint (URL). It sends an HTTP 302 status code to the client, which tells the client to go to another location. It is useful for navigation and to direct the user to another part of the application after certain actions, such as form submissions or authentication.


Q4. What are templates in Flask? Why is the render_template() function used?
Templates in Flask:

Templates in Flask are HTML files that contain static content and placeholders for dynamic content.
They allow you to separate the presentation (HTML structure) from your Python code (application logic).
Templates can contain variables and control structures (loops, conditions) that are replaced with actual values and logic when rendered.
render_template() function:

The render_template() function in Flask is used to render HTML templates.
It takes the name of the template file and additional variables as arguments.
This function allows you to pass data from your Python code to the HTML templates for dynamic rendering.

Let's create a simple Flask application that serves as an API, which will respond with JSON data. We'll use Postman to test the API.
from flask import Flask,request ,render_template , jsonify

app = Flask(__name__)


@app.route('/')
def home_page():
    return render_template('index.html')


@app.route('/math',methods=['POST'])
def math_ops():
    if(request.method == 'POST'):
        ops = request.form['operation']
        num1 = int(request.form['num1'])
        num2 = int(request.form['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'subtract':
            r = num1-num2
            result = "The subtract of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'multiply':
            r = num1*num2
            result = "The multiply of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'divide':
            r = num1/num2
            result = "The divide of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
            
        return render_template('results.html' , result = result)
